### Scatter Plot:
Use: Scatter plots are used to visualize the relationship between two numerical variables. Each data point is represented as a point on the plot, making it easy to see patterns and correlations.
Histogram:

### Histogram
Use: Histograms are useful for visualizing the distribution of a single numerical variable. They help you understand the data's central tendency, spread, and skewness.
Bar Plot:

### Bar Plot
Use: Bar plots are effective for displaying and comparing categorical data. They show the frequency or count of data points in each category, making it easy to compare different categories.
Box Plot (Box-and-Whisker Plot):

### Box Plot
Use: Box plots are used to visualize the distribution of numerical data and identify potential outliers. They display the median, quartiles, and potential outliers in the data, providing a summary of its distribution.
Heatmap:

### Heatmaps
Use: Heatmaps are used to display the relationships between two categorical variables or a numerical variable and a categorical variable. They use colors to represent the strength or frequency of a relationship, making them useful for exploring correlations and patterns in large datasets.
